```python
import pandas as pd

```

```python
b=pd.read_csv("May-2022.csv")
b.head(1)   #  2022 product info
```

```python
# Find rows in Transfer_Price with non-numeric values
invalid_transfer_price = b[~b['Transfer_Price'].str.replace(',', '').str.isnumeric()]


# Find rows in Final_Old_MRP with non-numeric values
count_nill = (b['Final_Old_MRP'].str.strip().str.lower() == 'nill').sum()
print("Number of rows with 'Nill' in Final_Old_MRP:", count_nill)
invalid_final_old_mrp = b[~b['Final_Old_MRP'].str.replace(',', '').str.isnumeric()]
print("Rows with invalid Final_Old_MRP:")
print(invalid_final_old_mrp)
```

```python
# Remove rows where Final_Old_MRP == 'Nill'
b = b[b['Final_Old_MRP'].str.strip().str.lower() != 'nill']

# Remove rows where Transfer_Price == '#VALUE!'
b = b[b['Transfer_Price'].str.strip() != '#VALUE!']

# Check results
print(b.shape)
print(b.head())
```

```python
unique_weights = b['Product_Weight'].unique()
print(unique_weights)
count_nill = (b['Product_Weight'].str.strip().str.lower() == 'nill').sum()
print("Number of rows with 'Nill' in Product_Weight:", count_nill)
invalid_product_weight = b[~b['Product_Weight'].str.replace(',', '').str.isnumeric()]
print("Rows with invalid Product_Weight:")
print(invalid_product_weight)
b = b.replace('Nill', 0)

```

```python


# Assuming your original DataFrame is named df
columns_to_remove = [
    'index','Old_MRP', 'Ajio_MRP', 'Amazon_MRP', 'Amazon_FBA_MRP', 
    'Flipkart_MRP', 'Limeroad_MRP', 'Myntra_MRP', 'Paytm_MRP', 'Snapdeal_MRP'
]

# Assign df to b first, then drop columns
c = b.drop(columns=columns_to_remove)

# Check the result
print(c.head())
```

```python
# Convert to float
c['Transfer_Price'] = c['Transfer_Price'].astype(float)
c['Final_Old_MRP'] = c['Final_Old_MRP'].astype(float)

# Convert to float with 2 decimals
c['Product_Weight'] = c['Product_Weight'].astype(float).round(2)

# Convert to string/text
text_cols = ['Product_SKU', 'Product_Style_ID', 'Product_Catalog', 'Product_Category']
c[text_cols] = c[text_cols].astype(str)

# Check the result
print(c.dtypes)
```

```python
c = c.rename(columns={
    'Product_SKU': 'sku',
    'Product_Style_ID': 'style',
    'Product_Catalog': 'catalog',
    'Product_Category': 'category',
    'Product_Weight': 'weight',
    'Transfer_Price': 'cost_price',
    'Final_Old_MRP': 'mrp'
})

# Check result
print(c.head())
```

```python
c.to_csv('new_2022_product_info.csv', index=False)
```
